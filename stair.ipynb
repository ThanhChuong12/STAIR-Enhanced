```python
import os
import shutil

os.chdir('/kaggle/working')

print("Đang dọn dẹp và tải mã nguồn...")
if os.path.exists('STAIR'):
    shutil.rmtree('STAIR')

os.system('git clone https://github.com/yhhe2004/STAIR.git')

# Cài đặt công cụ đo GPU và các thư viện Deep Learning lõi của STAIR
print("Đang cài đặt freerec, torch_geometric và pynvml...")
os.system('pip install nvidia-ml-py')  # thay pynvml đã deprecated
os.system('pip install torchdata==0.6.1 --no-deps')  # version còn datapipes
os.system('pip install freerec==0.8.3')              # version tương thích
os.system('pip install torch_geometric')

os.chdir('/kaggle/working/STAIR')

os.makedirs('data/baby', exist_ok=True)
os.makedirs('data/sports', exist_ok=True)
os.makedirs('data/electronics', exist_ok=True)

def extract_data(keyword, dest_dir):
    input_path = '/kaggle/input'
    found_files = 0
    for root, _, files in os.walk(input_path):
        if keyword.lower() in root.lower():
            for file in files:
                if file.endswith(('.npy', '.txt')):
                    src_file = os.path.join(root, file)
                    dest_file = os.path.join(dest_dir, file)
                    shutil.copy(src_file, dest_file)
                    found_files += 1
    
    if found_files > 0:
        print(f"  [+] Đã copy thành công {found_files} file cho tập: {keyword.capitalize()}")
    else:
        print(f"  [!] LỖI: Không tìm thấy dữ liệu cho tập {keyword.capitalize()}. Hãy kiểm tra lại phần Upload.")

print("Đang copy dữ liệu...")
extract_data('baby', 'data/baby')
extract_data('sports', 'data/sports')
extract_data('electronics', 'data/electronics')

print("\n✅ Đã thiết lập xong môi trường và copy an toàn 3 tập dữ liệu gốc!")
```

```python
import os, shutil

# root="../../data" relative to /kaggle/working/STAIR/main.py
# → /kaggle/working/STAIR/../../data → /kaggle/data/
DATA_ROOT = '/kaggle/data'

dataset_map = {
    'Amazon2014Baby_550_MMRec':        '/kaggle/input/datasets/rainyle/stair-datasets-mmrec/Amazon2014Baby_550_MMRec',
    'Amazon2014Sports_550_MMRec':      '/kaggle/input/datasets/rainyle/stair-datasets-mmrec/Amazon2014Sports_550_MMRec',
    'Amazon2014Electronics_550_MMRec': '/kaggle/input/datasets/rainyle/stair-datasets-mmrec/Amazon2014Electronics_550_MMRec',
}

for ds_name, src in dataset_map.items():
    dest = os.path.join(DATA_ROOT, ds_name)
    if os.path.exists(dest):
        shutil.rmtree(dest)
    shutil.copytree(src, dest)
    files = os.listdir(dest)
    print(f"✅ {ds_name}: {files}")

print("\n✅ Data ready!")
```

```python
import pynvml, time, threading, subprocess, matplotlib.pyplot as plt

all_logs = {ds: {'vram': [], 'time': []} for ds in ['baby', 'sports', 'electronics']}
profiling_active = False
current_dataset = None

def hardware_profiler(interval=0.5):
    pynvml.nvmlInit()
    handle = pynvml.nvmlDeviceGetHandleByIndex(0)
    start_time = time.time()
    while profiling_active:
        mem = pynvml.nvmlDeviceGetMemoryInfo(handle)
        if current_dataset:
            all_logs[current_dataset]['vram'].append(mem.used / (1024**2))
            all_logs[current_dataset]['time'].append(time.time() - start_time)
        time.sleep(interval)
    pynvml.nvmlShutdown()

# Tên dataset đúng như trong repo + hyperparams từ paper
runs = [
    {
        'key': 'baby',
        'dataset': 'Amazon2014Baby_550_MMRec',
        'weight_decay': '0.3',
        'gamma': '0.1',
        'batch_size': '1024',
    },
    {
        'key': 'sports',
        'dataset': 'Amazon2014Sports_550_MMRec',
        'weight_decay': '0.1',
        'gamma': '0.2',
        'batch_size': '1024',
    },
    {
        'key': 'electronics',
        'dataset': 'Amazon2014Electronics_550_MMRec',
        'weight_decay': '0.1',
        'gamma': '0.4',
        'batch_size': '4096',
    },
]

for run in runs:
    key = run['key']
    print(f"\n{'='*50}\n🚀 ĐANG HUẤN LUYỆN: {key.upper()}\n{'='*50}")

    current_dataset = key
    profiling_active = True
    t = threading.Thread(target=hardware_profiler, args=(0.5,))
    t.start()

    result = subprocess.run(
        [
            'python', 'main.py',
            '--dataset',      run['dataset'],
            '--root',         '/kaggle/data',
            '--weight-decay', run['weight_decay'],
            '--gamma',        run['gamma'],
            '--batch-size',   run['batch_size'],
            '--epochs',       '500',
            '--eval-freq',    '5',
        ],
        capture_output=True, text=True,
        cwd='/kaggle/working/STAIR'
    )

    profiling_active = False
    t.join()

    if result.returncode != 0:
        print(f"❌ LỖI {key.upper()}:")
        # In 4000 ký tự cuối stderr để thấy lỗi thực sự
        print(result.stderr[-4000:])
    else:
        print(f"✅ Hoàn thành {key.upper()}!")
        print(result.stdout[-1500:])

# ── Vẽ biểu đồ ──────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('GPU VRAM Usage — STAIR Training', fontsize=16, fontweight='bold')
colors = ['#1f77b4', '#ff7f0e', '#d62728']
keys = [r['key'] for r in runs]

for idx, (key, color) in enumerate(zip(keys, colors)):
    ax = axes[idx // 2, idx % 2]
    times, vrams = all_logs[key]['time'], all_logs[key]['vram']
    if vrams:
        ax.plot(times, vrams, color=color)
        ax.fill_between(times, vrams, color=color, alpha=0.3)
        ax.axhline(max(vrams), color='black', linestyle=':', label=f'Peak: {max(vrams):.1f} MB')
        ax.legend()
    ax.set_title(key.capitalize())
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('VRAM (MB)')
    ax.grid(True, linestyle=':', alpha=0.6)

ax_sum = axes[1, 1]
for key, color in zip(keys, colors):
    vrams = all_logs[key]['vram']
    if vrams:
        ax_sum.plot(all_logs[key]['time'], vrams, color=color,
                    label=f"{key} (Peak: {max(vrams):.1f} MB)")
ax_sum.set_title('Tổng hợp 3 datasets')
ax_sum.set_xlabel('Time (s)')
ax_sum.set_ylabel('VRAM (MB)')
ax_sum.grid(True, linestyle=':', alpha=0.6)
ax_sum.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('/kaggle/working/vram_profile.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n🎉 Hoàn thành toàn bộ thực nghiệm!")
```

```python
import pynvml
import time
import threading
import matplotlib.pyplot as plt
import subprocess

# Danh sách lưu trữ log cho từng dataset
all_logs = {
    'baby': {'vram': [], 'time': []},
    'sports': {'vram': [], 'time': []},
    'electronics': {'vram': [], 'time': []}
}

profiling_active = False
current_dataset = None

def hardware_profiler(interval=0.5): 
    pynvml.nvmlInit()
    handle = pynvml.nvmlDeviceGetHandleByIndex(0)
    start_time = time.time()
    
    while profiling_active:
        mem_info = pynvml.nvmlDeviceGetMemoryInfo(handle)
        if current_dataset:
            all_logs[current_dataset]['vram'].append(mem_info.used / (1024**2)) # Đổi ra MB
            all_logs[current_dataset]['time'].append(time.time() - start_time)
        time.sleep(interval)
    pynvml.nvmlShutdown()

# Đã sửa thành electronics
datasets_to_run = ['baby', 'sports', 'electronics']

for ds in datasets_to_run:
    print(f"\n{'='*50}\n🚀 ĐANG HUẤN LUYỆN: {ds.upper()}\n{'='*50}")
    
    # Kích hoạt profiler cho từng tập
    current_dataset = ds
    profiling_active = True
    profiler_thread = threading.Thread(target=hardware_profiler, args=(0.5,))
    profiler_thread.start()
    
    # Chạy lệnh huấn luyện với cờ --dataset
    subprocess.run(['python', 'main.py', '--dataset', ds])
    
    # Tắt profiler khi train xong 1 tập
    profiling_active = False
    profiler_thread.join()
    print(f"✅ Hoàn thành huấn luyện {ds.upper()}!")

# ==========================================
# TRỰC QUAN HÓA (VẼ 4 BIỂU ĐỒ)
# ==========================================
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Đo lường Tiêu thụ GPU VRAM (MB) trong quá trình huấn luyện STAIR', fontsize=16, fontweight='bold')

colors = ['#1f77b4', '#ff7f0e', '#d62728'] # Thay màu xanh lá bằng đỏ cho Electronics

# Vẽ 3 Biểu đồ riêng lẻ
for idx, (ds, color) in enumerate(zip(datasets_to_run, colors)):
    ax = axes[idx // 2, idx % 2]
    times = all_logs[ds]['time']
    vrams = all_logs[ds]['vram']
    
    if len(vrams) > 0:
        ax.plot(times, vrams, color=color, label=f'{ds.capitalize()}')
        ax.fill_between(times, vrams, color=color, alpha=0.3)
        ax.axhline(y=max(vrams), color='black', linestyle=':', label=f'Peak: {max(vrams):.1f} MB')
    
    ax.set_title(f'Tập dữ liệu: {ds.capitalize()}')
    ax.set_xlabel('Thời gian (giây)')
    ax.set_ylabel('VRAM (MB)')
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend(loc='lower right')

# Vẽ Biểu đồ tổng hợp (Góc dưới cùng bên phải)
ax_sum = axes[1, 1]
for ds, color in zip(datasets_to_run, colors):
    times = all_logs[ds]['time']
    vrams = all_logs[ds]['vram']
    if len(vrams) > 0:
        ax_sum.plot(times, vrams, color=color, label=f'{ds.capitalize()} (Peak: {max(vrams):.1f} MB)')

ax_sum.set_title('So sánh Tổng hợp 3 tập dữ liệu')
ax_sum.set_xlabel('Thời gian (giây)')
ax_sum.set_ylabel('VRAM (MB)')
ax_sum.grid(True, linestyle=':', alpha=0.6)
ax_sum.legend(loc='lower right')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

print("\n🎉 Đã hoàn thành toàn bộ thực nghiệm!")
```
